```python
cont = 1
numeros = []
tamanho_lista = int(input('Digite quantos numeros deseja '))
while cont <= tamanho_lista:
  numeros.append(cont)
  cont+=1
numeros
```

```python
pares = []
impares = []
for x in numeros:
  if x%2 == 0:
    pares.append(x)
  else:
    impares.append(x)
pares, impares
   
```

```python
cidades = []
inserir = input('Digite o nome da cidade:')
while inserir != '0':
  cidades.append(inserir)
  inserir = input('Digite o nome da cidade:')
for cidade in cidades:
  print(cidade)
```
