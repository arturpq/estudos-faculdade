```python
qtd_alunos = (int(input('Quantos alunos? ')))
alunos = []
notas = []
nota = 0
aprovados = []
reprovados = []
for x in range(qtd_alunos):
  alunos.append(input('Nome do aluno '))
  notas.append(int(input('Nota do aluno ')))
for aluno in alunos:
  print(aluno,':',notas[nota])
  if notas[nota] >=7:
    aprovados.append(aluno)
  else:
    reprovados.append(aluno) 
  nota+=1
if len(aprovados) != 0:
  print('Os aprovados foram')
  for num in range(len(aprovados)):
    print(aprovados[num])
else:
  print('Não houveram aprovados')
if len(reprovados) !=0:
  print('Os reprovados foram')
  for num in range(len(reprovados)):
    print(reprovados[num])
  print('Tendo um total de',len(aprovados),'Aprovados e',len(reprovados),'Reprovados')
else:
  print('Não houveram reprovados')
```
