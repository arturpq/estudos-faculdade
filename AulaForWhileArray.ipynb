```python
alunos = []
quantidade_nomes = int(input('Quantos nomes deseja inserir? '))
print_nomes = 0
while quantidade_nomes > len(alunos):
  alunos.append(str(input('Insira o nome ')))

while print_nomes < quantidade_nomes:
  print (alunos[print_nomes])
  print_nomes = print_nomes +1
```

```python
aluno_retirado = str(input('Qual nome deseja retirar da lista? '))
check_retirado = 0
if aluno_retirado in alunos:
  while aluno_retirado != alunos[check_retirado]:
    check_retirado = check_retirado + 1
  else:
    del alunos[check_retirado]  
  print ('O nome '+ aluno_retirado + ' foi retirado da lista')
else:
  print ('O nome '+ aluno_retirado + ' não está na lista e não pode ser retirado')
```

```python
aluno_pedido = str(input('Qual nome deseja ver se tem na lista? '))
if aluno_pedido in alunos:
  print ('O nome '+ aluno_pedido + ' está na lista')
else:
  print ('O nome '+ aluno_pedido + ' não está na lista')
```
